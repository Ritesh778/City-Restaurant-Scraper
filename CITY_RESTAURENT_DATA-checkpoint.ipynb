```python
import pandas as pd  #imported for data manipulation and data analysis.
import time #library is imported for adding delays in the code, to make sure that the website has enough time to load.
from selenium import webdriver #webdriver is a module from selenium library, which provides a way to control a web browser through the code.
from bs4 import BeautifulSoup # library for pulling data out of HTML and XML files. It is used for web scraping.
from urllib.parse import urljoin   #used to join a base URL and a relative URL to form a complete URL.
from selenium.webdriver.common.by import By  #By function is from the selenium.webdriver.common library, and is used to specify a mechanism to locate elements within a web page.
headers = {
    'authority': 'scrapeme.live',
    'dnt': '1',
    'upgrade-insecure-requests': '1',
    'user-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.61 Safari/537.36',
    'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,/;q=0.8,application/signed-exchange;v=b3;q=0.9',
    'sec-fetch-site': 'none',
    'sec-fetch-mode': 'navigate',
    'sec-fetch-user': '?1',
    'sec-fetch-dest': 'document',
    'accept-language': 'en-GB,en-US;q=0.9,en;q=0.8',
}  # this is use to tell zomato website from which source or which user is accessing their website because zomato use secure network

driver = webdriver.Chrome(executable_path='C:\selenium drivers\chromedriver.exe')  # this is selenium driver which is use to automate the website
driver.get("https://www.zomato.com/visakhapatnam/vizag-city-order-online")  # zomato URL for getting data
time.sleep(2)  # time dealy for wait to open website
scroll_pause_time = 3  # scroll time  by which zomato website auto scroll after each 3 second
screen_height = driver.execute_script("return window.screen.height;")  # this is use to define screen height
i = 1

while True:  # run a loop until website reach end of the page

    driver.execute_script("window.scrollTo(0, {screen_height}*{i});".format(screen_height=screen_height,i=i))  # its execute scroll positions
    i += 1
    time.sleep(scroll_pause_time)

    scroll_height = driver.execute_script("return document.body.scrollHeight;")

    if (screen_height) * i > scroll_height:
        break  # when website reach end of the page then while loop break

soup = BeautifulSoup(driver.page_source, "html.parser")  # this is use to get the data from the page open above
divs1 = soup.findAll('div',class_='sc-1mo3ldo-0')  # this is a class in zomato website. by this class we get div html tag. under this tag all required data is listed

divs=divs1[4:]     #The line divs=divs1[4:] is slicing the divs1 list to only include the elements starting from the fourth index. This is done to exclude any irrelevant information that may be present in the first few elements of the list.
print(len(divs1))
print(len(divs1[4:]))
# create arr to store data for further use
urls = []
rest_name = []
ratings = []
addresses = []
price = []
crusine = []
df=pd.DataFrame()
# run a loop until we found child div tag. in these child div tag we get data from each restaurant

for parent in divs:
    try:
        isCom=True
        try:
            if "Inspiration for your first order" in parent.find("h3").text: #The code starts with a for loop that iterates over a list of divs. In each iteration, the code tries to extract the relevant information by checking the text contained in the different HTML tags present within the div element.
                print("no need first")
                isCom = False
        except Exception as e:
            print("")

        try:
            if "Best Food in" in parent.find("h1").text: #the code checks if the text of the h1 tag contains the phrase "Best Food in" and the text of the h3 tag contains the phrase "Top brands for". If either of these phrases is present, the code sets the isCom variable to False.
                print("no need best")
                isCom = False
        except Exception as e:
            print("")
        
        try:
            if "Top brands for" in parent.find("h3").text: 
                print("no need top")
                isCom = False
        except Exception as e:
            print("")
        try:
            if "End of search results" in parent.find("h3").text: #If the text of the h3 tag contains the phrase "End of search results", the code sets the isCom variable to False and breaks the for loop.
                print("no need end")
                isCom = False
                break
        except Exception as e:
            print("")
        if isCom:  #After checking the text of the different HTML tags, if the value of the isCom variable is True, the code goes ahead to extract the URLs of the restaurants. The code first finds all the div elements with the class jumbo-tracker, and iterates over them.
            rits = parent.findAll('div',class_='jumbo-tracker')
            print(len(rits))
            i =0
            for par in rits:

                print(i)
                i=i+1
                link_tag = par.find("a")  # here we get link of that restaurant of zomato page

                base = "https://www.zomato.com"

                if 'href' in link_tag.attrs:   #In each iteration, the code finds the a tag within the div element, and retrieves its href attribute. The code then combines the href attribute with the base URL of the Zomato website to form the full URL of the restaurant. This URL is then appended to the urls list.
                    link = link_tag.get('href')

                url = urljoin(base, link)  # not we join base zomato link & restaurant link and add in url array
                urls.append(url)   
                print(url)

                rating_tag = par.div.a.next_sibling.div.div.div.div.div.div.div.text
                price_tag = par.div.a.next_sibling.p.next_sibling.text
                crusine_tag = par.div.a.next_sibling.p.text
                ratings.append(rating_tag)
                price.append(price_tag)
                crusine.append(crusine_tag)

                driver.get(url)
                time.sleep(3)
                soup = BeautifulSoup(driver.page_source, "html.parser")  # this is use to get the data from the page open above

                cuisine_name = []
                prices = []
                description = []
                rest_names = []
                rest_names.append(soup.select('h1.sc-7kepeu-0')[-1].text)
                rest_name.append(soup.select('h1.sc-7kepeu-0')[-1].text)
                for h in soup.find_all('h4', class_="sc-1s0saks-15 iSmBPS"):
                    cuisine_name.append(h.text)

                rest_names = rest_names * len(cuisine_name)  #The rest_names list is multiplied by the length of cuisine_name list, to ensure that the same restaurant name is repeated for each cuisine.

                for h1 in soup.find_all('span', class_="sc-17hyc2s-1 cCiQWA"):
                    prices.append(h1.text)

                for h2 in soup.find_all('p', class_="sc-1s0saks-12 hcROsL"):
                    description.append(h2.text)

                if df.empty:
                    df = pd.DataFrame(zip(rest_names, cuisine_name, prices, description))
                else:
                    df1 = pd.DataFrame(zip(rest_names, cuisine_name, prices, description))
                    df = df.append(df1, ignore_index=True)

                df.to_csv('Zomato1.csv', index=False)

    except Exception as e:
        print(e)
        pass


df3=pd.DataFrame({'rest_names':rest_name,'rest_urls':urls})
df2 = pd.DataFrame({'rest_urls': urls,'rest_ratings': ratings, 'price for two': price,'rest_crusine': crusine})# we create csv data
df4=pd.merge(left=df3, right=df2,how='inner',on='rest_urls')
df4.to_csv("restaurants_data2.csv")  # save csv data in  file
driver.close()
```
